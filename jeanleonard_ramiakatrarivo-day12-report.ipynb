### Exercise 1:
Create a simple line plot for `y = x` for `x` values ranging from 0 to 10.

### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

```python
x = np.linspace(-10, 10, 400)

y1 = x**2
y2 = x**3

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Plot y1 = x^2 in the first subplot
axs[0].plot(x, y1, color='blue')
axs[0].set_title('Plot of $y = x^2$')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

# Plot y2 = x^3 in the second subplot
axs[1].plot(x, y2, color='orange')
axs[1].set_title('Plot of $y = x^3$')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

# Adjust layout
plt.tight_layout()

plt.show()
```

### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.

```python

x = np.linspace(-10, 10, 400)

y1 = x
y2 = x**2
y3 = x**3

# Create the plot
plt.figure(figsize=(10, 6))

# Plot y1 = x 
plt.plot(x, y1, color='blue', linestyle='-', label='$y = x$')

# Plot y2 = x^2 
plt.plot(x, y2, color='red', linestyle='--', label='$y = x^2$')

# Plot y3 = x^3 
plt.plot(x, y3, color='green', linestyle=':', label='$y = x^3$')

plt.title('Plot of Three Functions')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
```

### Exercise 11:
Create a pie chart with 5 segments labeled A, B, C, D, and E, with values `[25, 35, 20, 15, 5]`.

### Exercise 14:
Plot a filled area under the curve of `y = x^2` for `x` ranging from 0 to 10.

```python
x = np.linspace(0, 10, 100)
y = x**2 

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label='$y = x^2$')
plt.fill_between(x, y, color='skyblue', alpha=0.5) 

plt.title('Filled Area Under the Curve of $y = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

### Exercise 16:
Create a stacked bar chart for the following data: `Group A = [3, 5, 7]`, `Group B = [2, 6, 4]`, and `categories = ['X', 'Y', 'Z']`.

```python
categories = ['X', 'Y', 'Z']
group_A = [3, 5, 7]
group_B = [2, 6, 4]

x = np.arange(len(categories))

plt.figure(figsize=(8, 6))
plt.bar(x, group_A, label='Group A', color='skyblue')
plt.bar(x, group_B, bottom=group_A, label='Group B', color='salmon')

plt.title('Stacked Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(x, categories) 
plt.legend()

plt.show()
```

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.

```python
x = np.linspace(0, 10, 100)

y1 = x
y2 = x**2
y3 = x**3
y4 = x**4

# Create subplots in a 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot y1 = x in the first subplot
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('$y = x$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

# Plot y2 = x^2 in the second subplot
axs[0, 1].plot(x, y2, color='red')
axs[0, 1].set_title('$y = x^2$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

# Plot y3 = x^3 in the third subplot
axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('$y = x^3$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

# Plot y4 = x^4 in the fourth subplot
axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_title('$y = x^4$')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')

plt.tight_layout()

plt.show()
```

### Exercise 18:
Create a heatmap using a 2D array of random values with shape (5, 5).

```python
data = np.random.rand(5, 5)

# Create the heatmap
plt.figure(figsize=(6, 6))
plt.imshow(data, cmap='viridis', aspect='auto')
plt.colorbar(label='Value')

plt.title('Heatmap of Random Values')
plt.xticks(np.arange(5), ['A', 'B', 'C', 'D', 'E'])
plt.yticks(np.arange(5), ['1', '2', '3', '4', '5'])

plt.show()
```

### Exercise 19:
Add error bars to a bar plot with `categories = ['A', 'B', 'C', 'D']` and `values = [3, 7, 5, 9]`, assuming the error is 1 for each value.

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]
errors = [1] * len(values)  # Error of 1 for each value

plt.figure(figsize=(8, 6))
plt.bar(categories, values, yerr=errors, color='skyblue', capsize=5)
plt.title('Bar Plot with Error Bars')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y')

plt.show()
```
